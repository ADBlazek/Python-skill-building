```
# All imports here
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from numpy.random import randint
from numpy import mean
from scipy import stats
from numpy.random import seed
```

```
# Created a line chart for normal distribution; plotted 1000 trials of sample
# mean

results = []
for num_trials in range(1,1000):
    trials = np.random.normal(1, 7, num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})


df.plot()
```

```
# Created a line chart for gamma distribution; plotted 1000 trials of sample
# mean
results = []
for num_trials in range(1,1000):
    trials = np.random.gamma(1, 7, num_trials)
    mean_of_trials = trials.mean()
    results.append(mean_of_trials)
	
df = pd.DataFrame({ 'trials' : results})


df.plot()
```

```
# Normal histogram of 1000 randomly sampled means
means = [mean(np.random.normal(1, 7, 1000)) for _ in range(1000)]


plt.hist(means)
plt.show()

print(stats.describe(means))
```

```
# Gamma histogram of 1000 randomly sampled means
means = [mean(np.random.gamma(1, 7, 1000)) for _ in range(1000)]


plt.hist(means)
plt.show()

print(stats.describe(means))
```
