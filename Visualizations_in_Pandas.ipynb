```
# Imports here
import pandas as pd
import seaborn as sns 
sns.set()

# Suppress warnings (warnings do not prevent the code from executing)
import warnings
warnings.filterwarnings("ignore")

# Download statistics about houseprices as a DataFrame
houseprices = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/houseprices.csv')
```

```
# A look at the DataFram
houseprices.head()
```

```
# Another look
houseprices.info()
```

```
# Illustrate the distribution of saleprice.
# Descriptive stats here
houseprices['saleprice'].describe()
```

```
# A look at the visual distribution
ax = sns.distplot(houseprices['saleprice'])
```

```
# Illustrate the distribution of the lotarea.
houseprices['lotarea'].describe()
```

```
ax = sns.distplot(houseprices['lotarea'])
```

```
# Use a boxplot to illustrate how overallcond affects the sale price.
sns.boxplot(y="saleprice", x="overallcond", data=houseprices);
# As overall condition increases, seems to lead to an increase in sale prices
```

```
# On the same boxplot, highlight the difference between having central air
# conditioning or not.
ax = sns.boxplot(y="saleprice", hue="centralair", x="overallcond", data=houseprices);
# Central air increases sale prices across all conditions.
```

```
# Is there a correlation between lotarea and the saleprice?
sns.scatterplot(x="lotarea", y="saleprice", data=houseprices);
# It looks like there is a positive correlation, but I will
# run stats to be sure.
```

```
# Stats show there is a statistically significant positive correlation.
from scipy import stats
stats.pearsonr(houseprices['lotarea'], houseprices['saleprice'])
```
