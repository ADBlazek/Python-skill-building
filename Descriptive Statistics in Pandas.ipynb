```
# Download statistics about houseprices as a DataFrame
import pandas as pd
houseprices = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/houseprices.csv')
```

```
# Are there any nulls?

# Yes, several columns contain null values.

houseprices.info()
```

```
# How many distinct values are in the mszoning column?
houseprices['mszoning'].nunique()
```

```
# How many distinct values are in the street column?
houseprices['street'].nunique()
```

```
# Use the describe method on the range salesprice.  
# What is the minimum value in the range? $34,900

houseprices['saleprice'].describe()
```

```
# Use the describe method on the range lotarea.  
# What is the mean value in the range? $10,517

houseprices['lotarea'].describe()
```

```
# Is there a statistically significant difference between the prices of houses 
# that have an open porch versus houses that do not have an open porch? How did
# you handle null values? Hint: use the variable openporchsf. You will need to 
# transform this variable so that it can be used in a t-test.

houseprices[['openporchsf', 'saleprice']].info()

# There are no null values in the columns of interest. A zero in openporchsf
# means that there is no open porch. 
```

```
# Creating new dataframes of interest and getting information about them.
openporches = houseprices[['openporchsf', 'saleprice']].iloc[(houseprices['openporchsf'] > 0).values]
openporches.info()
```

```
no_openporches = houseprices[['openporchsf', 'saleprice']].iloc[(houseprices['openporchsf'] == 0).values]
no_openporches.info()
```

```
# T-test for signif between the two above conditions
from scipy import stats
stats.ttest_ind(openporches['saleprice'], no_openporches['saleprice'])

# The p value is less than 0.05, so the test is significant. There is a difference
# in sale prices between houses with and without open porches.
```

```
# Is there a significant correlation between the lotarea and the saleprice variables?
# Are they positively or negatively correlated?

# Yes, these variables are significantly postively correlated.

stats.pearsonr(houseprices['lotarea'], houseprices['saleprice'])
```
